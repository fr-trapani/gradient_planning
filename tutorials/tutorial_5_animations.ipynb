```python
from matplotlib import pyplot as plt

from agents.pgp.pgp_softmax import SoftMaxPGP
from environments.gridworlds.gridworlds_classic import FourRoom
from plots.gridworlds.gridworld_visualizer import GridWorldVisualizer
from plots.gridworlds.gridworld_animator import GridWorldAnimator
from utils.policy_functions import *
```

```python
# Maze Parameters
maze_type = FourRoom
maze_size = 5

# Agent Parameters
gamma = 0.99
p_func = p0_onehot

# Training Parameters
n_steps = 1000
alpha = 0.02
gradient_kwargs = {"natural":False}

# Figure Parameters
n_frames = 200
```

```python
world = maze_type(maze_size)
agent = SoftMaxPGP(world)
viz = GridWorldVisualizer(world, agent)
ani = GridWorldAnimator(viz)
```

```python
viz.plot_gradient_norm()
plt.tight_layout()
plt.show()

viz.plot_gradient()
plt.tight_layout()
plt.show()

viz.plot_policy()
plt.tight_layout()
plt.show()
```

```python
# Training
agent.reset()
agent.learn(n_steps=n_steps, alpha=alpha, gradient_kwargs=gradient_kwargs)

viz.plot_gradient_norm(policy_gradient=True)
plt.tight_layout()
plt.show()

viz.plot_gradient()
plt.tight_layout()
plt.show()

viz.plot_policy()
plt.tight_layout()
plt.show()

viz.plot_trajectory_distribution()
plt.tight_layout()
plt.show()
```

```python
a = ani.animate_policy_divergence(ts_interval=round(n_steps/n_frames), fps=20)
a.ipython_display(fps=20, loop=False, autoplay=True)
```
